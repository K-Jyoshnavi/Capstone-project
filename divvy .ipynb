```python
import math
from math import radians, sin, cos, sqrt, atan2
#Haversine Formula
def haversine(lat1, lon1, lat2, lon2):
    # Radius of the Earth in kilometers
    R = 6371.0

    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Calculate the differences between latitudes and longitudes
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # Calculate the distance
    distance = R * c

    return distance
```

```python
# Calculate distance and store the result in a new column
all_data['distance_travelled_km'] = all_data.apply(lambda row: haversine(row['start_lat'], row['start_lng'], row['end_lat'], row['end_lng']), axis=1)
all_data.head()
```

```python
date_columns = ['started_at','ended_at']
all_data[date_columns] = all_data[date_columns].apply(pd.to_datetime)

#Calculate ride duration and store the result in a new column
all_data['ride_duration_s']=(all_data['ended_at']-all_data['started_at']).dt.total_seconds()
all_data.head()
```

```python
#Filter out ride duration less than 1 minute
all_data_filtered = all_data[all_data['ride_duration_s'] >= 60]
all_data =all_data_filtered.copy(deep=True)
all_data.head()
     
```

```python
#Calculate speed in kph
all_data['speed_kph'] = (all_data['distance_travelled_km'] / (all_data['ride_duration_s'] / 3600))

#Check the spread of data of column 'speed_kph'
all_data['speed_kph'].describe()
```

```python
#Add 'day_of_week' and 'month' based on 'started_at'
#df_add = df05_filtered.copy(deep=True)
all_data['day_of_week']=all_data['started_at'].dt.strftime('%A')
all_data['month']=all_data['started_at'].dt.strftime('%B')
all_data.head()
```

```python
def day_cat (d):
    if d == "Saturday" or d == "Sunday":
        return "Weekend"
    else: 
        return "Weekday"
     
```

```python
#Add 'day_type'
all_data['day_type'] = all_data['day_of_week'].apply(day_cat)
all_data.head()
     
```

```python
def szn (m):
    if m == "December" or m == "January" or m == "February":
        return "Winter"
    elif m == "March" or m == "April" or m == "May":
        return "Spring"
    elif m == "June" or m == "July" or m == "August":
        return "Summer"
    else: 
        return "Fall"
```

```python
#Add 'season'
all_data['season'] = all_data['month'].apply(szn)
all_data.head()
     
```

```python
#Function for 'route_type'
def rte_ty (row):
    if row['start_station_name'] == row['end_station_name']:
        return "Round trip"
    else: 
        return "One-way trip"
```

```python
#Add 'route_type'
all_data['route_type'] = all_data.apply(rte_ty, axis=1)
all_data.head()
```

```python
#Add 'ride_duration_min'
all_data ['ride_duration_min'] = all_data['ride_duration_s'] / 60
all_data.head()
```

```python
df_format = all_data.copy(deep=True)
df_format.rename(columns = {'rideable_type':'bike_type','started_at':'start_time', 'ended_at':'end_time', 'member_casual':'user_type' }, inplace = True)
new_cols = ['bike_type', 'user_type','start_time','end_time','day_of_week','day_type','month','season','start_station_name','end_station_name','route_type', 'start_lat', 'start_lng', 'end_lat','end_lng', 'distance_travelled_km', 'ride_duration_s', 'ride_duration_min']
all_data=df_format[new_cols]
all_data.head()
```
